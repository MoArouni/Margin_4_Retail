```python
import numpy as np
import matplotlib.pyplot as plt

# Number of items
n = 200

# Define the range for buying prices (B) and selling prices (S)
B = np.linspace(3, 10, 100)  # Buying prices from £5 to £10
S = np.linspace(15, 25, 100)  # Selling prices from £20 to £30

# Create a meshgrid for plotting
B, S = np.meshgrid(B, S)

# Calculate total margin using the formula
M_total = n * (S - B)

# Create a contour plot
plt.figure(figsize=(10, 6))
contour = plt.contourf(S, B, M_total, levels=50, cmap='viridis')
plt.colorbar(contour, label='Total Margin (£)')

plt.title('Total Margin Contour Plot')
plt.xlabel('Selling Price (S)')
plt.ylabel('Buying Price (B)')

# Set x and y ticks
plt.xticks(np.arange(15, 25, 0.6))  # Set x ticks with 0.6 spacing
plt.yticks(np.arange(3, 10, 0.25))   # Set y ticks with 0.2 spacing

plt.grid(True)
plt.show()
```

```python
def calculate_margin_percentage(selling_price, buying_price, quantity):
    """Calculate the margin percentage."""
    total_cost = buying_price * quantity
    total_revenue = selling_price * quantity
    margin = total_revenue - total_cost
    margin_percentage = (margin / total_revenue) * 100
    return margin_percentage

def main():
    while True:
        # User input for items, selling price, and buying price
        try:
            quantity = int(input("Enter the amount of items being bought and sold: "))
            buying_price = float(input("Enter the buying price per item: £"))
            selling_price = float(input("Enter the selling price per item: £"))
            
            # Calculate the margin percentage
            margin_percentage = calculate_margin_percentage(selling_price, buying_price, quantity)
            print(f"\nCurrent margin percentage: {margin_percentage:.2f}%\n")

            # User input for target margin
            target_margin = float(input("Enter your target margin percentage: "))
            
            # Calculate required adjustments
            if target_margin > margin_percentage:
                a = input("Would you like to see the SP change (type S) or the BP (type B) to reach your TM:")
                if a == "S":
                    x = round((buying_price*quantity * 100) / (100 - target_margin), 2)
                    new_selling_price = round(x/quantity, 2)
                    print(f"Your new selling price to meet your target margin is {new_selling_price}")
                    new_margin_percentage = calculate_margin_percentage(new_selling_price, buying_price, quantity)
                    print(f"\nCurrent margin percentage: {new_margin_percentage:.2f}%\n")
                elif a == "B": 
                    y = round((((selling_price*quantity*100) - (selling_price*quantity*target_margin)) / 100) , 2)
                    new_buying_price = round(y/quantity, 2)
                    print(f"Your new buying price to meet your target margin is {new_buying_price}")
                    new_margin_percentage = calculate_margin_percentage(selling_price, new_buying_price, quantity)
                    print(f"\nCurrent margin percentage: {new_margin_percentage:.2f}%\n")
            else:
                print("Your current margin percentage already meets your target margin.")

        except ValueError:
            print("Invalid input. Please enter numeric values.")
        
        # Ask if the user wants to continue
        continue_prompt = input("\nDo you want to calculate again? (yes/no): ").strip().lower()
        if continue_prompt != 'yes':
            break

if __name__ == "__main__":
    main()


```
